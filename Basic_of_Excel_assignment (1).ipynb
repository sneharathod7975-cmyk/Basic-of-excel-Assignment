 Q1.) **Explain the difference between Relative , Absolute and Mixed Cell Referencing.**

 You have the following dataset:
  
  -> 1. **Relative cell reference**:- a Relative cell reference is a defauld cell reference inspreadsheet programs like Excel that automatically adjusts when a formula is copied to defferent cell

  . Example= A1+B1 when you copy this formula  down or across, Excel automatically changes the cell referance relative to theire new position.

  Example :(Formula) C1=A1+B1 C2(copied down)=A2+B2 C3(copied down again)=A3+B3 so, the reference move along with the formula.

 -> 2. **Absolute cell reference** :-locks a cell's reference so it does not change when copied to other cells

  . example:= $A$1 + $B$1
  
  . When you copy this formula anywere, both column A and row 1 stay fixed- they don't  change.
  
  Example: Formula C1=$A$1+ $B$1 C2(COPIED DOWN)=$A$1+$B$1 D5(copied anywhere)=$A$1+$B$1

->3.** Mixed cell Reference**:- A mixed cell reference combines an absolute (fixed) and a relative (changing) reference,Using a dollar sing ($) to lock either the column or the row

 . Example : $A-> Column fixed, row change

 .A$1 -> Row fixed, column changes Example: $A1- fixes Column A, row changes, useful when copying across columns A$1- fixes row

 1. column changes, Useful when copying down rows
